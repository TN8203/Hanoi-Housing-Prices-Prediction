Most apartments have house direction and balcony direction as 1 of 4 sub directions (South-east, North-east, South-west, North-west)

```python
direction_mapping = {
    'Đông': 'Hướng chính',
    'Tây': 'Hướng chính',
    'Nam': 'Hướng chính',
    'Bắc': 'Hướng chính',
    'Đông - Bắc': 'Hướng phụ',
    'Đông - Nam': 'Hướng phụ',
    'Tây - Bắc': 'Hướng phụ',
    'Tây - Nam': 'Hướng phụ'
}

data['Nhóm hướng nhà'] = data['Hướng nhà'].map(direction_mapping)

avg_price_by_housedirection = data[data['Mức giá']>10].groupby('Nhóm hướng nhà')['Mức giá'].agg(['mean', 'max', 'min'])
avg_price_by_housedirection
```

```python
data['Nhóm hướng ban công'] = data['Hướng ban công'].map(direction_mapping)

# Tính mức giá trung bình theo nhóm hướng nhà
avg_price_by_balconydirection = data[data['Mức giá']>10].groupby('Nhóm hướng ban công')['Mức giá'].agg(['mean', 'max', 'min'])
avg_price_by_balconydirection
```

Although the main and sub directions have a difference in quantity, they do not affect the house price

Chọn hướng mặt quay về Nam hay Đông Nam ( đón ánh nắng vào ban ngày, mùa đông tránh được gió thổi từ phía Bắc)
Tránh ban công cửa sổ quay về hướng Tây, Tây Bắc: phía Tây là phía đón gió nóng vào mua hè nên tránh những chung cư có ban công quay về phía Tây và Bắc)

```python
direction_weather_mapping = {
    'Đông': 'Bình thường',
    'Tây': 'Thời tiết không thuận lợi',
    'Nam': 'Thời tiết thuận lợi',
    'Bắc': 'Thời tiết không thuận lợi',
    'Đông - Bắc': 'Bình thường',
    'Đông - Nam': 'Thời tiết thuận lợi',
    'Tây - Bắc': 'Thời tiết không thuận lợi',
    'Tây - Nam': 'Bình thường'
}

data['Nhóm hướng ban công'] = data['Hướng ban công'].map(direction_weather_mapping)

avg_price_by_balconydirection = data[data['Mức giá']>10].groupby('Nhóm hướng ban công')['Mức giá'].agg(['mean', 'max', 'min'])
avg_price_by_balconydirection
```

```python
select_col = ['Nhóm 1-Trường học', 'Nhóm 1-Chợ/Siêu thị',
              'Nhóm 1-Bệnh viện', 'Nhóm 2-Công viên', 'Nhóm 2-Bể bơi', 'Nhóm 2-Gym',
              'Slot ô tô', 'Nhà thoáng', 'Hiện đại', 'Bán gấp', 'Lô góc',
              'Hỗ trợ vay', 'Tầng']

# Convert the selected columns to numeric, coercing errors
data[select_col] = data[select_col].apply(pd.to_numeric, errors='coerce')

# Calculate the sums
data['Tổng số tiện ích'] = data[select_col].sum(axis=1).astype(int)
data['Tiện ích cơ bản'] = data[['Nhóm 1-Trường học', 'Nhóm 1-Chợ/Siêu thị', 'Nhóm 1-Bệnh viện']].sum(axis=1)
data['Tiện ích giải trí'] = data[['Nhóm 2-Công viên', 'Nhóm 2-Bể bơi', 'Nhóm 2-Gym']].sum(axis=1)
```
